```python
log_dir = "/home/hqyone/mnt/2tb/github/cancer_rcnn/data/logs"
weights_path = "/home/hqyone/mnt/2tb/github/cancer_rcnn/code/cell_classification/model/mask_rcnn_coco.h5"
#weights_path="coco"

dataset_dir="/home/hqyone/mnt/2tb/github/cancer_rcnn/data"
subset="stage1_train" 

def hpv_train(log_dir, weights_path, dataset_dir, subset):
    config = NucleusConfig()
    model = modellib.MaskRCNN(mode="training", config=config, model_dir=log_dir)
    model.load_weights(weights_path, by_name=True, exclude=[
            "mrcnn_class_logits", "mrcnn_bbox_fc",
            "mrcnn_bbox", "mrcnn_mask"])
    train(model, dataset_dir, subset, config)

def predict(weights_path, dataset_dir, subset,result_dir):
    config = NucleusInferenceConfig()
    model = modellib.MaskRCNN(mode="inference", config=config, model_dir=log_dir)
    model.load_weights(weights_path, by_name=True)
    detect(model,dataset_dir,subset,result_dir)

#hpv_train(log_dir, weights_path, dataset_dir, subset)
```
